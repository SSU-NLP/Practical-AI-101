### ⚡ 왜 텐서(벡터화 연산)를 써야 하나요?

텐서를 사용하는 **가장 큰 이유** 중 하나는 속도입니다.
Python의 `for` 반복문으로 배열 원소를 하나씩 처리하는 것보다,
PyTorch의 벡터화 연산을 사용하면 **내부적으로 병렬 처리**되어 훨씬 빠릅니다.

`%%timeit` 매직 명령어로 실제 속도 차이를 비교해봅시다.

> 💡 **참고:** NumPy도 같은 이유로 벡터화를 권장하며, PyTorch 텐서도 동일한 원리가 적용됩니다.
> GPU 환경에서는 그 차이가 훨씬 더 극적으로 나타납니다.


**방법 1: Python `for` 반복문으로 원소별 연산 (느림 🐢)**


```python
%%timeit
x = torch.rand(1000, 1000)
for i in range(100, 1000):
    for j in range(x.shape[1]):
        x[i, j] += 5
```

**방법 2: PyTorch 벡터화 연산으로 한번에 처리 (빠름 🚀)**


```python
%%timeit
x = torch.rand(1000, 1000)
x[torch.arange(100, 1000), :] += 5
```

수십 배에서 수백 배 이상의 속도 차이가 나는 것을 확인할 수 있습니다.

딥러닝에서 텐서를 다룰 때 **절대 원소별 `for` 루프를 사용하지 마세요.**
`.sum()`, `.mean()`, `.max()`, 슬라이싱, 브로드캐스팅 등 PyTorch가 제공하는
벡터화 연산으로 항상 대체할 수 있습니다.


### 브로드캐스팅 (Broadcasting)

서로 **모양(shape)이 다른** 텐서끼리도 연산이 될 때가 있습니다.
PyTorch가 작은 텐서를 자동으로 "늘려서(broadcast)" 큰 텐서 모양에 맞춰주기 때문입니다.

딥러닝에서는 이게 정말 자주 쓰입니다. 예를 들어 **신경망의 편향(bias)을 더하거나**,
**배치(batch) 전체에 같은 값을 적용**할 때 브로드캐스팅이 내부적으로 동작합니다.


### 🎬 시각 자료: 브로드캐스팅 (Broadcasting)

크기가 1인 차원이 실제로 늘어나(stretch) 같은 모양이 된 뒤 원소별로 더해집니다.

![Broadcasting](assets/Broadcasting.gif)

> ManimGL로 제작 · gif는 같은 저장소의 assets/ 에서 로드됩니다 (GitHub·로컬용)


```python
# (3, 4) 행렬의 각 행(row)에 (4,) 벡터를 더해봅시다
matrix = torch.arange(12, dtype=torch.float32).reshape(3, 4)
bias = torch.tensor([100., 200., 300., 400.])

print("matrix:")
print(matrix)
print("bias:", bias)
print("\nmatrix + bias  (bias가 모든 행에 자동으로 더해짐):")
print(matrix + bias)
```

```python
# (3, 1) 과 (1, 4) 를 더하면 -> (3, 4) 로 확장됩니다
col = torch.tensor([[1.], [2.], [3.]])      # 세로 벡터 (3, 1)
row = torch.tensor([[10., 20., 30., 40.]])  # 가로 벡터 (1, 4)

result = col + row
print(result)
print("결과 shape:", result.shape)  # (3, 4)

# 💡 규칙: 뒤쪽 차원부터 비교해서, 크기가 같거나 둘 중 하나가 1이면 OK
```

## Part 2: 자동 미분 (Autograd)

PyTorch의 핵심 기능 중 하나는 **자동 미분(Automatic Differentiation)**입니다.
`backward()` 메서드를 호출하면 PyTorch가 자동으로 그래디언트를 계산하고, 이를 `grad` 속성에 저장합니다.

![sgd](https://blog.kakaocdn.net/dna/rXDcn/btrIuYYVC1u/AAAAAAAAAAAAAAAAAAAAAIZYSwWIGzvQm-4eb1UHb_PqszfGe7Hx6b6uS5uxxZ_m/img.gif?credential=yqXZFxpELC7KVnFOS48ylbz2pIh7yKj8&expires=1782831599&allow_ip=&allow_referer=&signature=9MrUSFKv5tUm2ZOFrbtY5ry3p3U%3D)

![backpropagation](https://media.geeksforgeeks.org/wp-content/uploads/20240217152156/Frame-13.png)

*역전파(Backpropagation): 출력에서 발생한 오차를 각 레이어로 거꾸로 흘려보내면서 가중치의 그래디언트를 구해내는 과정입니다.*

### 🎬 시각 자료: 자동 미분 · 역전파 (Backpropagation)

순전파(파랑)는 입력→은닉→출력→손실로 흐르고, backward()는 기울기(빨강)를 손실→출력→은닉→입력으로 거꾸로 흘려 각 층의 가중치 기울기를 연쇄 법칙으로 구합니다.

![Backprop](assets/Backprop.gif)

> ManimGL로 제작 · gif는 같은 저장소의 assets/ 에서 로드됩니다 (GitHub·로컬용)


```python
# Create an example tensor
# requires_grad parameter tells PyTorch to store gradients
x = torch.tensor([2.], requires_grad=True)

# Print the gradient if it is calculated
# Currently None since x is a scalar
pp.pprint(x.grad)
```

```python
# Calculating the gradient of y with respect to x
y = x * x * 3 # 3x^2
y.backward()
pp.pprint(x.grad) # d(y)/d(x) = d(3x^2)/d(x) = 6x = 12
```

다른 텐서에서 역전파(backprop)를 다시 실행하면 어떻게 되는지 살펴봅시다:


```python
z = x * x * 3 # 3x^2
z.backward()
pp.pprint(x.grad)
```

```python
x.grad = None
z = x * x * 3 # 3x^2
z.backward()
# y = x * x * 3
pp.pprint(x.grad)
```

```python
z = x * x * 3 # 3x^2
z.backward()
# y = x * x * 3
pp.pprint(x.grad)
```

```python
z = x * x * 3 # 3x^2
z.backward()
# y = x * x * 3
pp.pprint(x.grad)
```

`x.grad`는 지금까지 계산된 그래디언트의 **누적합**으로 업데이트됩니다.
신경망 학습 시, 특정 뉴런에 대한 모든 그래디언트를 합산한 뒤 파라미터를 업데이트합니다.
바로 이 이유 때문에 매 학습 반복(iteration)마다 `zero_grad()`를 호출해야 합니다.
그렇지 않으면 그래디언트가 계속 누적되어 잘못된 업데이트가 발생합니다.

### 선형 레이어 (Linear Layer)

`nn.Linear(H_in, H_out)`을 사용하면 선형 레이어를 만들 수 있습니다.
`(N, *, H_in)` 형태의 입력을 받아 `(N, *, H_out)` 형태의 출력을 반환합니다 (`*`는 임의 개수의 중간 차원).
선형 레이어는 `Ax + b` 연산을 수행하며, `A`와 `b`는 랜덤으로 초기화됩니다.
편향(bias)을 학습하지 않으려면 `bias=False`로 초기화하면 됩니다.


### 🎬 시각 자료: 선형 레이어 = 뉴런 = Wx+b

nn.Linear가 입력의 가중합+편향이라는 점, 그리고 그것이 곧 행렬 연산 Wx+b임을 보여줍니다.

### 레이어 연결하기 (Putting the Layers Together)

```python
block = nn.Sequential(
    nn.Linear(4, 2),
    nn.Sigmoid()
)

input = torch.ones(2,3,4)
output = block(input)
output
```

### 커스텀 모듈 (Custom Modules)

미리 정의된 모듈 외에도 `nn.Module` 클래스를 상속하여 직접 모듈을 만들 수 있습니다.

커스텀 모듈 작성 방법:
1. `nn.Module`을 상속합니다.
2. `__init__` 함수에서 레이어들을 초기화합니다 (반드시 `super().__init__()` 먼저 호출).
3. `nn` 모듈 객체로 정의된 클래스 속성들은 학습 가능한 파라미터로 처리됩니다.
   일반 텐서는 `nn.Parameter`로 감싸면 파라미터가 됩니다.
4. `forward(x)` 함수를 반드시 구현해야 합니다. `model(x)` 호출 시 이 함수가 실행됩니다.


### 🎬 시각 자료: MLP 순전파 (Forward Pass)

전체 파이프라인(Linear→ReLU→Linear→Sigmoid)을 한눈에 보여준 뒤 각 단계를 하이라이트하며 최종 출력이 어떻게 나오는지 따라갑니다.

![MLPForward](assets/MLPForward.gif)

> ManimGL로 제작 · gif는 같은 저장소의 assets/ 에서 로드됩니다 (GitHub·로컬용)


```python
class MultilayerPerceptron(nn.Module):

  def __init__(self, input_size, hidden_size):
    # Call to the __init__ function of the super class
    super(MultilayerPerceptron, self).__init__()

    # Bookkeeping: Saving the initialization parameters
    self.input_size = input_size
    self.hidden_size = hidden_size

    # Defining of our model
    # There isn't anything specific about the naming of `self.model`. It could
    # be something arbitrary.
    self.model = nn.Sequential(
        nn.Linear(self.input_size, self.hidden_size),
        nn.ReLU(),
        nn.Linear(self.hidden_size, self.input_size),
        nn.Sigmoid()
    )

  def forward(self, x):
    output = self.model(x)
    return output
```

`nn.Sequential` 대신 각 레이어를 `__init__`에서 개별 정의하고,
`forward`에서 직접 연결하는 방식으로도 동일하게 구현할 수 있습니다:


```python
class MultilayerPerceptron(nn.Module):

  def __init__(self, input_size, hidden_size):
    # Call to the __init__ function of the super class
    super(MultilayerPerceptron, self).__init__()

    # Bookkeeping: Saving the initialization parameters
    self.input_size = input_size
    self.hidden_size = hidden_size

    # Defining of our layers
    self.linear = nn.Linear(self.input_size, self.hidden_size)
    self.relu = nn.ReLU()
    self.linear2 = nn.Linear(self.hidden_size, self.input_size)
    self.sigmoid = nn.Sigmoid()

  def forward(self, x):
    linear = self.linear(x)
    relu = self.relu(linear)
    linear2 = self.linear2(relu)
    output = self.sigmoid(linear2)
    return output
```

클래스를 정의했으니, 이제 인스턴스를 생성하고 실행해봅시다:


```python
# Make a sample input
input = torch.randn(2, 5)

# Create our model
model = MultilayerPerceptron(5, 3)

# Pass our input through our model
model(input)
```

`named_parameters()`와 `parameters()` 메서드로 모델의 파라미터를 확인할 수 있습니다:


```python
list(model.named_parameters())
```

## Part 4: 최적화 (Optimization)

`backward()` 함수로 그래디언트를 계산하는 방법을 살펴보았습니다.
하지만 그래디언트만으로는 모델이 학습되지 않습니다.
파라미터를 어떻게 업데이트할지 결정하는 **옵티마이저(Optimizer)**가 필요합니다.

`torch.optim` 모듈에는 다양한 옵티마이저가 있습니다. 대표적인 예로 `optim.SGD`와 `optim.Adam`이 있습니다.
옵티마이저를 초기화할 때 `model.parameters()`를 전달하여 어떤 값을 최적화할지 알려줍니다.
학습률(`lr`)은 매 step에서 업데이트 크기를 결정합니다.


### 🎬 시각 자료: 경사 하강법 (Gradient Descent)

L(x)=3x² 위의 공이 기울기(6x)의 반대 방향으로 한 걸음씩 내려가 최솟값에 수렴합니다.

![GradientDescent](assets/GradientDescent.gif)

> ManimGL로 제작 · gif는 같은 저장소의 assets/ 에서 로드됩니다 (GitHub·로컬용)


```python
import torch.optim as optim
```

옵티마이저를 정의한 후, 최적화할 **손실 함수(Loss Function)**를 정의합니다.
직접 정의하거나 PyTorch의 미리 정의된 함수(`nn.BCELoss()` 등)를 사용할 수 있습니다.
모든 것을 합쳐봅시다! 먼저 더미 데이터를 생성합니다:


```python
# Create the y data
y = torch.ones(10, 5)

# Add some noise to our goal y to generate our x
# We want out model to predict our original data, albeit the noise
x = y + torch.randn_like(y)
x
```

이제 모델, 옵티마이저, 손실 함수를 정의합니다:


```python
# Instantiate the model
model = MultilayerPerceptron(5, 3)

# Define the optimizer
adam = optim.Adam(model.parameters(), lr=1e-1)

# Define loss using a predefined loss function
loss_function = nn.MSELoss()

# Calculate how our model is doing now
y_pred = model(x)
loss_function(y_pred, y).item()
```

손실을 더 줄일 수 있는지 확인해봅시다. 이제 **학습 루프(Training Loop)**를 구성합니다:


### 🎬 시각 자료: 학습 루프 · 옵티마이저 (Training Loop)

zero_grad→forward→loss→backward→step 사이클을 돌며 w ← w - η∇ 로 파라미터를 갱신하고, 에폭마다 손실이 줄어듭니다.

![TrainingLoop](assets/TrainingLoop.gif)

> ManimGL로 제작 · gif는 같은 저장소의 assets/ 에서 로드됩니다 (GitHub·로컬용)


```python
# Set the number of epoch, which determines the number of training iterations
n_epoch = 10

for epoch in range(n_epoch):
  # Set the gradients to 0
  adam.zero_grad()

  # Get the model predictions
  y_pred = model(x)

  # Get the loss
  loss = loss_function(y_pred, y)

  # Print stats
  print(f"Epoch {epoch}: traing loss: {loss}")

  # Compute the gradients
  loss.backward()

  # Take a step to optimize the weights
  adam.step()
```

## 1. 데이터 준비

먼저 학습 데이터가 필요합니다. 여기서는 짧은 영화 리뷰 문장과 그 감정 레이블을 직접 만들어 사용합니다.

**레이블: `1` = 긍정, `0` = 부정**

> 실제 프로젝트에서는 IMDB, 네이버 영화 리뷰(NSMC) 같은 수만 개짜리 데이터셋을 쓰지만,
> 원리는 똑같습니다. 여기서는 빠른 실습을 위해 작은 데이터로 시작합니다.


```python
train_data = [
    ("this movie was fantastic and i loved it", 1),
    ("an amazing film with brilliant acting", 1),
    ("what a wonderful and touching story", 1),
    ("the best movie i have seen this year", 1),
    ("absolutely loved the characters and the plot", 1),
    ("a great film truly enjoyable and fun", 1),
    ("the acting was superb and the story was beautiful", 1),
    ("i really enjoyed every minute of it", 1),
    ("such a delightful and charming movie", 1),
    ("brilliant direction and a fantastic ending", 1),
    ("a heartwarming and excellent film", 1),
    ("the plot was clever and the acting wonderful", 1),
    ("this was a terrible and boring movie", 0),
    ("an awful film with bad acting", 0),
    ("what a waste of time and money", 0),
    ("the worst movie i have ever seen", 0),
    ("i hated the characters and the plot", 0),
    ("a horrible film completely disappointing", 0),
    ("the acting was bad and the story was dull", 0),
    ("i did not enjoy any part of it", 0),
    ("such a boring and forgettable movie", 0),
    ("poor direction and a terrible ending", 0),
    ("a dreadful and painful film", 0),
    ("the plot was stupid and the acting awful", 0),
]

print(f"전체 문장 수: {len(train_data)}")
print("긍정 예시:", train_data[0])
print("부정 예시:", train_data[-1])
```

### 문장을 숫자로 바꾸기 (토큰화 + 어휘집)

신경망은 글자가 아니라 **숫자**만 다룰 수 있습니다. 그래서 두 단계가 필요합니다.

1. **토큰화(Tokenization):** 문장을 단어 단위로 쪼갭니다. (`"i loved it"` → `["i", "loved", "it"]`)
2. **어휘집(Vocabulary):** 모든 단어에 고유한 번호(인덱스)를 부여합니다.

특수 토큰 두 개도 추가합니다:
- `<pad>` (인덱스 0): 문장 길이를 맞추기 위한 빈칸 채우기용
- `<unk>`: 어휘집에 없는(처음 보는) 단어를 처리하기 위한 토큰

![Tokenization](https://camo.githubusercontent.com/937931bc39d9211a7d16803fcea8ce621422a635f72934af703468774e30bcd7/68747470733a2f2f73656261737469616e72617363686b612e636f6d2f696d616765732f4c4c4d732d66726f6d2d736372617463682d696d616765732f636830325f636f6d707265737365642f30342e77656270)

```python
# 1) 토큰화: 소문자로 바꾸고 공백 기준으로 분리
def tokenize(sentence):
    return sentence.lower().split()

# 2) 어휘집 만들기
vocab = set()
for sentence, _ in train_data:
    vocab.update(tokenize(sentence))

# 특수 토큰을 맨 앞에 두고, 나머지는 정렬해서 추가
vocab = ["<pad>", "<unk>"] + sorted(vocab)

# 단어 <-> 인덱스 사전
word_to_ix = {word: i for i, word in enumerate(vocab)}
ix_to_word = {i: word for word, i in word_to_ix.items()}

print(f"어휘집 크기: {len(vocab)}")
print("'fantastic'의 인덱스:", word_to_ix["fantastic"])
print("앞쪽 단어 10개:", vocab[:10])
```

### 인덱스로 변환하고 길이 맞추기 (패딩)

이제 각 문장을 인덱스 리스트로 바꿉니다. 그런데 문장마다 길이가 다르죠.

텐서는 직사각형 모양이어야 하므로, 짧은 문장 뒤에 `<pad>`(인덱스 0)를 채워 **길이를 통일**합니다.


```python
# 문장 -> 인덱스 리스트 (없는 단어는 <unk>로)
def encode(sentence):
    return [word_to_ix.get(tok, word_to_ix["<unk>"]) for tok in tokenize(sentence)]

# 가장 긴 문장 길이에 맞춰 뒤를 <pad>(0)로 채우기
max_len = max(len(encode(s)) for s, _ in train_data)
print("가장 긴 문장 길이:", max_len)

def pad(ids):
    return ids + [word_to_ix["<pad>"]] * (max_len - len(ids))

# 전체 데이터를 텐서로 변환
X = torch.tensor([pad(encode(s)) for s, _ in train_data], dtype=torch.long)
y = torch.tensor([label for _, label in train_data], dtype=torch.float32).unsqueeze(1)

print("X(입력) shape:", X.shape, " <- (문장 수, 문장 길이)")
print("y(레이블) shape:", y.shape)
print("\n첫 문장의 인덱스 표현:")
print(train_data[0][0])
print(X[0])
```

## 2. 모델 설계 — 임베딩 평균 풀링

모델 구조는 아주 단순하지만 강력합니다:

1. **임베딩(`nn.Embedding`):** 각 단어 인덱스를 의미를 담은 벡터로 변환합니다.
   (Part 3에서 자세히 다루지만, 여기서 미리 써봅니다.)
2. **평균 풀링(Mean Pooling):** 문장 속 단어 벡터들을 **평균**내어 "문장 전체를 대표하는 벡터" 하나를 만듭니다.
   (`<pad>`는 평균에서 제외합니다.)
3. **선형 분류기(`nn.Linear`):** 그 벡터를 받아 긍정일 점수(logit) 하나를 출력합니다.

> 💡 `padding_idx=0`을 주면 `<pad>` 토큰의 임베딩이 학습되지 않고 0으로 유지됩니다.


### 🎬 시각 자료: 평균 풀링 (Mean Pooling)

단어마다 '분위기 화살표'가 생기고, 그걸 다 평균 내면 문장 전체의 무게중심 = 전반적 분위기가 나옵니다. 여러 명 별점을 평균 내 한 줄 평을 만드는 것과 같죠. 길이가 3단어든 30단어든 항상 벡터 1개로 요약돼 길이 문제도 해결됩니다. (단, 평균은 순서를 무시해서 '별로 안 좋아' 같은 문장은 놓칠 수 있음 → 이후 attention으로 이어집니다.)

![MeanPooling](assets/MeanPooling.gif)

> ManimGL로 제작 · gif는 같은 저장소의 assets/ 에서 로드됩니다 (GitHub·로컬용)


```python
import torch.nn as nn

class SentimentClassifier(nn.Module):
    def __init__(self, vocab_size, embed_dim=24, pad_ix=0):
        super().__init__()
        # 단어 인덱스 -> embed_dim 차원 벡터
        self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=pad_ix)
        # 문장 벡터 -> 긍정 점수 1개
        self.fc = nn.Linear(embed_dim, 1)
        self.pad_ix = pad_ix

    def forward(self, x):
        # x: (배치, 문장길이)
        emb = self.embedding(x)                  # (배치, 길이, embed_dim)

        # <pad> 토큰은 평균에서 제외 (mask로 가림)
        mask = (x != self.pad_ix).unsqueeze(-1)  # (배치, 길이, 1)
        summed = (emb * mask).sum(dim=1)         # 실제 단어들만 합
        counts = mask.sum(dim=1).clamp(min=1)    # 실제 단어 개수
        sentence_vec = summed / counts           # 평균 풀링 -> (배치, embed_dim)

        return self.fc(sentence_vec)             # (배치, 1) logit

# 모델 생성
model = SentimentClassifier(vocab_size=len(vocab))
print(model)
```

## 3. 학습 (Training)

이제 Part 1에서 배운 **학습 루프**를 그대로 사용합니다.

- **손실 함수:** `nn.BCEWithLogitsLoss` — 이진 분류(긍정/부정)에 쓰는 표준 손실 함수입니다.
  (내부적으로 sigmoid를 적용하므로 모델은 logit만 출력하면 됩니다.)
- **옵티마이저:** `Adam`
- 매 epoch마다 `zero_grad → forward → loss → backward → step` 순서를 반복합니다.


```python
loss_fn = nn.BCEWithLogitsLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.05)

losses = []
n_epoch = 150
for epoch in range(n_epoch):
    optimizer.zero_grad()          # 1) 그래디언트 초기화
    logits = model(X)              # 2) 예측 (forward)
    loss = loss_fn(logits, y)      # 3) 손실 계산
    loss.backward()                # 4) 그래디언트 계산 (backward)
    optimizer.step()               # 5) 파라미터 업데이트
    losses.append(loss.item())

    if (epoch + 1) % 30 == 0:
        with torch.no_grad():
            preds = (torch.sigmoid(model(X)) > 0.5).float()
            acc = (preds == y).float().mean().item()
        print(f"Epoch {epoch+1:3d} | loss {loss.item():.4f} | 정확도 {acc*100:.0f}%")
```

```python
# 손실이 줄어드는 과정을 그래프로 확인
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.plot(losses)
plt.title("Training Loss (lower is better)")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.grid(alpha=0.3)
plt.show()
```

## 5. 보너스 — 임베딩이 학습한 *의미* 들여다보기 🔍

신기한 부분입니다. 우리는 단어의 의미를 알려준 적이 없는데도,
모델은 학습 과정에서 **비슷한 감정의 단어들을 비슷한 벡터로** 만들어 냅니다.

`embedding` 레이어에 학습된 벡터들 사이의 **코사인 유사도(cosine similarity)** 를 계산해
어떤 단어들이 서로 가까운지 확인해 봅시다.


```python
import torch.nn.functional as F
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 학습된 임베딩 벡터들 (어휘집 크기, embed_dim)
emb = model.embedding.weight.data

def most_similar(word, topn=5):
    if word not in word_to_ix:
        print(f"'{word}'는 어휘집에 없는 단어예요.")
        return
    vec = emb[word_to_ix[word]]
    sims = F.cosine_similarity(emb, vec.unsqueeze(0), dim=1)
    order = sims.argsort(descending=True).tolist()
    order = [i for i in order if ix_to_word[i] not in ("<pad>", "<unk>", word)][:topn]

    print(f"'{word}'와(과) 의미가 비슷한 단어 Top {topn}:")
    for i in order:
        print(f"   {ix_to_word[i]:14s}  유사도 {sims[i].item():.2f}")
    print()

most_similar("fantastic")
most_similar("terrible")
```

```python
def show_pca():
  # 시각화할 단어들 선택
  words_to_plot = ["fantastic", "amazing", "wonderful", "best", "terrible", "awful", "boring", "worst"]
  indices = [word_to_ix[w] for w in words_to_plot if w in word_to_ix]
  word_labels = [ix_to_word[i] for i in indices]
  vectors = emb[indices].cpu().numpy()

  # 2차원 PCA
  pca = PCA(n_components=2)
  vectors_2d = pca.fit_transform(vectors)

  # 그래프 그리기
  plt.figure(figsize=(10, 8))
  plt.axhline(0, color='grey', lw=1)
  plt.axvline(0, color='grey', lw=1)

  # 원점에서 단어 벡터 방향으로 화살표 그리기
  for i, word in enumerate(word_labels):
      x, y = vectors_2d[i, 0], vectors_2d[i, 1]
      plt.arrow(0, 0, x, y, head_width=0.05, head_length=0.1, fc='blue', ec='blue', alpha=0.6)
      plt.text(x * 1.2, y * 1.2, word, fontsize=12, fontweight='bold')

  plt.title("Word Embedding Space (PCA 2D)")
  plt.xlabel("Principal Component 1")
  plt.ylabel("Principal Component 2")
  plt.grid(True, linestyle='--', alpha=0.5)
  plt.axis('equal')
  plt.show()

show_pca()
```

# Part 3: 실전 예제 ② — Word Window Classification

지금까지 PyTorch의 기초를 배우고 간단한 신경망을 만들어봤습니다.
이제 실제 NLP 과제를 처음부터 끝까지 풀어봅니다. 다룰 내용은 다음과 같습니다:

1. **데이터**: 배치 텐서(Batched Tensor)로 데이터셋 만들기

NLP 과제에서 코퍼스(corpus)는 보통 `.txt`나 `.csv` 파일이며, 각 행이 하나의 문장에 해당합니다.

```python
# 학습 코퍼스: (문장, 단어별 LOCATION 레이블)
# 핵심 — 같은 단어 'jordan'이 문맥에 따라 지역(1)이기도 사람(0)이기도 합니다.
# 그래서 단어만 보는 정적 규칙으로는 라벨을 못 만들고, 문맥(윈도우)이 필요합니다.
ww_data = [
    # --- jordan = 지역 (LOCATION) ---
    ("we traveled to jordan last summer",    [0, 0, 0, 1, 0, 0]),
    ("she flew to jordan for vacation",      [0, 0, 0, 1, 0, 0]),
    ("the capital of jordan is amman",       [0, 0, 0, 1, 0, 1]),
    ("i visited paris and jordan",           [0, 0, 1, 0, 1]),
    ("they will move to jordan soon",        [0, 0, 0, 0, 1, 0]),
    # --- jordan = 사람 (PERSON) ---
    ("michael jordan scored forty points",   [0, 0, 0, 0, 0]),
    ("jordan passed the ball quickly",       [0, 0, 0, 0, 0]),
    ("do you know jordan well",              [0, 0, 0, 0, 0]),
    ("jordan and michael played basketball", [0, 0, 0, 0, 0]),
    ("my friend jordan called me",           [0, 0, 0, 0, 0]),
    # --- 기타 지역 / 비지역 (균형용) ---
    ("the professor is from australia",      [0, 0, 0, 0, 1]),
    ("i live in stanford",                   [0, 0, 0, 1]),
    ("he comes from taiwan",                 [0, 0, 0, 1]),
    ("we always come to paris",              [0, 0, 0, 0, 1]),
    ("the capital of turkey is ankara",      [0, 0, 0, 1, 0, 1]),
]

corpus = [sentence for sentence, _ in ww_data]
print(f"문장 수: {len(corpus)}")
corpus
```

### 전처리 (Preprocessing)

모델 학습을 쉽게 하기 위해 데이터에 몇 가지 전처리 과정을 적용합니다.
특히 텍스트 데이터에서 중요합니다. 대표적인 전처리 기법들:
* **토큰화(Tokenization)**: 문장을 단어 단위로 분리
* **소문자화(Lowercasing)**: 모든 글자를 소문자로 변환
* **노이즈 제거(Noise removal)**: 특수문자(구두점 등) 제거
* **불용어 제거(Stop words removal)**: 자주 쓰이는 의미 없는 단어 제거

어떤 전처리가 필요한지는 과제에 따라 다릅니다.
이번 과제에서는 **소문자화**와 **토큰화**만 적용합니다.


```python
# 학습 예제를 생성하는 데 사용될 전처리 함수
# 이 함수는 간단하게 단어를 소문자로 바꾸고 토큰화합니다.
def preprocess_sentence(sentence):
  return sentence.lower().split()

# 학습 세트 생성
train_sentences = [preprocess_sentence(sent) for sent in corpus]
train_sentences
```

```python
# 단어별 레이블 (위 ww_data에서 그대로 가져옵니다)
# 같은 'jordan'이라도 문장에 따라 1(지역)/0(사람)이 다릅니다 — 문맥 의존 레이블.
train_labels = [labels for _, labels in ww_data]
train_labels
```

### 단어를 임베딩으로 변환하기 (Converting Words to Embeddings)

학습 데이터의 각 데이터포인트는 단어의 시퀀스입니다.
하지만 머신러닝 모델은 벡터 형태의 숫자로 동작합니다. 단어를 어떻게 숫자로 바꿀까요?

바로 **임베딩(embedding)** 입니다!

임베딩 룩업 테이블 `E`가 있다고 상상해봅시다. 각 행이 하나의 임베딩에 대응합니다.
단어의 임베딩을 찾는 과정은 다음과 같습니다:
1. 임베딩 테이블에서 단어의 인덱스 `i`를 찾기: `word → index`
2. 인덱스로 테이블에 접근해 임베딩 가져오기: `index → embedding`

먼저 첫 단계를 위해, 어휘집(vocabulary)의 모든 단어에 인덱스를 부여합니다:
1. 코퍼스의 모든 고유 단어 찾기
2. 각 단어에 인덱스 할당


### 🎬 시각 자료: 임베딩 룩업 (nn.Embedding)

nn.Embedding은 학습되는 룩업 테이블입니다 — 단어 → 인덱스 → 테이블의 해당 행 → 그 행이 곧 임베딩 벡터.

![EmbeddingLookup](assets/EmbeddingLookup.gif)

```python
# 코퍼스에 있는 모든 고유한 단어 찾기
vocabulary = set(w for s in train_sentences for w in s)
vocabulary
```

이제 `vocabulary`에는 코퍼스의 모든 단어가 담겼습니다.
하지만 테스트 시에는 어휘집에 없는 단어가 등장할 수 있습니다.

이를 처리하기 위해 특수 토큰 `<unk>`(unknown)을 도입합니다.
어휘집에 없는 단어(out-of-vocabulary)는 모두 이 토큰으로 표현합니다.
이 토큰을 어휘집에 추가합니다.


```python
# 어휘집에 알 수 없는(unknown) 토큰 추가
vocabulary.add("<unk>")
```

앞서 이 과제가 `Word Window Classification`이라 불리는 이유를 언급했습니다.
모델이 예측할 때 해당 단어뿐 아니라 **주변 단어**까지 함께 보기 때문입니다.

예를 들어 "We always come to Paris" 문장의 레이블은 `0, 0, 0, 0, 1`입니다 (Paris만 위치).
모델이 `Paris`에 대해 예측할 때, `Paris`만 보면 `to`가 위치와 자주 함께 등장한다는
중요한 정보를 놓치게 됩니다.

**윈도우(window)** 는 각 단어의 앞뒤 `±N`개 단어를 함께 고려하게 해줍니다.
그런데 문제가 있습니다. `Paris`는 문장 끝에 있어서 뒤따르는 단어가 없습니다.
PyTorch 모델은 입력 차원이 고정되어 있으므로, 윈도우 크기가 `1`이면 매번 `3`개 단어를 받아야 하는데
때때로 `2`개만 받을 수는 없습니다.

해결책은 특수 토큰 `<pad>`를 도입해 모든 단어가 유효한 윈도우를 갖도록 문장 양끝을 채우는 것입니다.


```python
# 어휘집에 <pad> 토큰 추가
vocabulary.add("<pad>")

# 주어진 문장을 패딩하는 함수
# 이 함수는 예시로 여기에 포함되었으며, 튜토리얼 후반부에 사용될 예정입니다.
def pad_window(sentence, window_size, pad_token="<pad>"):
  window = [pad_token] * window_size
  return window + sentence + window

# 패딩 예시 보여주기
window_size = 2
pad_window(train_sentences[0], window_size=window_size)
```

어휘집이 준비되었으니, 각 단어에 인덱스를 부여합니다.

> 💡 패딩 토큰의 인덱스를 `0`으로 두면 편리합니다.
> `nn.utils.rnn.pad_sequence` 같은 일부 PyTorch 함수가 기본값으로 0을 사용하기 때문입니다.

```python
# 토큰 문장이 주어지면 해당 인덱스를 반환합니다.
def convert_token_to_indices(sentence, word_to_ix):
  indices = []
  for token in sentence:
    # 토큰이 어휘집에 있는지 확인합니다. 있다면 해당 인덱스를 가져옵니다.
    # 없다면 알 수 없는(unknown) 토큰의 인덱스를 가져옵니다.
    if token in word_to_ix:
      index = word_to_ix[token]
    else:
      index = word_to_ix["<unk>"]
    indices.append(index)
  return indices

# 동일한 함수의 더 간결한 버전
def _convert_token_to_indices(sentence, word_to_ix):
  return [word_to_ix.get(token, word_to_ix["<unk>"]) for token in sentence]

# 예시 보여주기
example_sentence = ["we", "always", "come", "to", "kuwait"]
example_indices = convert_token_to_indices(example_sentence, word_to_ix)
restored_example = [ix_to_word[ind] for ind in example_indices]

print(f"원래 문장은: {example_sentence}")
print(f"단어에서 인덱스로: {example_indices}")
print(f"인덱스에서 단어로: {restored_example}")
```

### 문장 배치 처리 (Batching Sentences)

전체 코퍼스를 한 번에 처리한 뒤 업데이트하는 것은 비용이 큽니다.
반대로 매 예제마다 업데이트하면 손실(loss)이 불안정해집니다.
그래서 **배치(batch)** 단위로 학습한 뒤 파라미터를 업데이트합니다.
이렇게 하면 전체 손실의 그래디언트를 더 잘 추정할 수 있습니다.

`torch.utils.data.DataLoader` 클래스로 데이터를 배치로 구성하는 법을 배워봅시다.
`DataLoader(data, batch_size=..., shuffle=True, collate_fn=...)` 형태로 호출합니다.
* `batch_size`: 배치당 예제 수
* `shuffle=True`: 배치 순서를 섞어 같은 나쁜 배치를 반복해서 만나지 않도록 함
* `collate_fn`: 배치를 가공하는 커스텀 함수

우리의 `collate_fn`은 다음 작업을 수행합니다:
1. 학습 문장에 윈도우 패딩 적용
2. 단어를 인덱스로 변환
3. 한 배치 안의 모든 문장/레이블 길이를 동일하게 패딩
   (손실 계산 시 실제 단어 수가 필요하므로 길이도 기록)

`collate_fn`이 `word_to_ix` 딕셔너리에 접근해야 하므로, Python의 `partial` 함수를 활용합니다.


![model_architecture](assets/dataloader_collate_pipeline.png)

```python
from torch.utils.data import DataLoader
from functools import partial

def custom_collate_fn(batch, window_size, word_to_ix):
  # 배치를 훈련 예제(x)와 레이블(y)로 분리합니다.
  # nn.utils.rnn.pad_sequence 메서드가 텐서를 예상하기 때문에 x와 y를 텐서로 변환합니다.
  # 모델이 텐서 입력을 예상하기 때문에 이것도 유용합니다.
  x, y = zip(*batch)

  # 이제 훈련 예제를 윈도우 패딩해야 합니다. 윈도우 패딩을 처리하는 함수를 이미 정의했습니다.
  # 모든 것이 한 곳에 있도록 여기에 다시 포함합니다.
  def pad_window(sentence, window_size, pad_token="<pad>"):
    window = [pad_token] * window_size
    return window + sentence + window

  # 훈련 예제를 패딩합니다.
  x = [pad_window(s, window_size=window_size) for s in x]

  # 이제 훈련 예제의 단어를 인덱스로 변환해야 합니다.
  # 위와 동일한 이유로 이전에 정의된 함수를 복사합니다.
  def convert_tokens_to_indices(sentence, word_to_ix):
    return [word_to_ix.get(token, word_to_ix["<unk>"]) for token in sentence]

  # 훈련 예제를 인덱스로 변환합니다.
  x = [convert_tokens_to_indices(s, word_to_ix) for s in x]

  # 이제 배치의 모든 예제 길이가 동일하도록 예제를 패딩할 것입니다.
  # 이렇게 하면 행렬 연산이 가능해집니다.
  # batch_first 매개변수를 True로 설정하여 반환된 행렬이 배치를 첫 번째 차원으로 갖도록 합니다.
  pad_token_ix = word_to_ix["<pad>"]

  # pad_sequence 함수는 입력이 텐서여야 하므로 x를 텐서로 변환합니다.
  x = [torch.LongTensor(x_i) for x_i in x]
  x_padded = nn.utils.rnn.pad_sequence(x, batch_first=True, padding_value=pad_token_ix)

  # 레이블도 패딩할 것입니다. 그렇게 하기 전에, 각 예제에 얼마나 많은 단어가 있었는지 알 수 있도록
  # 레이블의 수를 기록할 것입니다.
  lengths = [len(label) for label in y]
  lengths = torch.LongTensor(lengths)

  y = [torch.LongTensor(y_i) for y_i in y]
  y_padded = nn.utils.rnn.pad_sequence(y, batch_first=True, padding_value=0)

  # 이제 변수를 반환할 준비가 되었습니다. 여기서 변수를 반환하는 순서는
  # 훈련 루프에서 읽는 순서와 일치합니다.
  return x_padded, y_padded, lengths
```

함수가 길어 보이지만, 사실 그렇게 길 필요는 없습니다.
아래는 보조 함수 선언과 주석을 제거한 간결한 버전입니다.


```python
def _custom_collate_fn(batch, window_size, word_to_ix):
  # 데이터 포인트 준비
  x, y = zip(*batch)
  x = [pad_window(s, window_size=window_size) for s in x]
  x = [convert_token_to_indices(s, word_to_ix) for s in x]

  # x를 패딩하여 배치의 모든 예제가 동일한 크기를 갖도록 합니다.
  pad_token_ix = word_to_ix["<pad>"]
  x = [torch.LongTensor(x_i) for x_i in x]
  x_padded = nn.utils.rnn.pad_sequence(x, batch_first=True, padding_value=pad_token_ix)

  # y를 패딩하고 길이를 기록합니다.
  lengths = [len(label) for label in y]
  lenghts = torch.LongTensor(lengths)
  y = [torch.LongTensor(y_i) for y_i in y]
  y_padded = nn.utils.rnn.pad_sequence(y, batch_first=True, padding_value=0)

  return x_padded, y_padded, lenghts
```

이제 `DataLoader`가 실제로 어떻게 동작하는지 살펴봅시다.


```python
# DataLoader에 전달될 매개변수
data = list(zip(train_sentences, train_labels))
batch_size = 2
shuffle = True
window_size = 2
collate_fn = partial(custom_collate_fn, window_size=window_size, word_to_ix=word_to_ix)

# DataLoader 인스턴스화
loader = DataLoader(data, batch_size=batch_size, shuffle=shuffle, collate_fn=collate_fn)

# 한 루프를 통과합니다.
counter = 0
for batched_x, batched_y, batched_lengths in loader:
  print(f"Iteration {counter}")
  print("Batched Input:")
  print(batched_x)
  print("Batched Labels:")
  print(batched_y)
  print("Batched Lengths:")
  print(batched_lengths)
  print("")
  counter += 1
```

위에서 본 배치 입력 텐서가 모델에 전달됩니다.
현재 입력은 한 문장의 모든 단어가 하나의 데이터포인트에 담긴 형태입니다.
모델은 이 입력에서 각 단어별 윈도우를 만들고, 중심 단어가 `LOCATION`인지 예측한 뒤,
예측들을 모아 반환해야 합니다.

윈도우 크기가 `N`이면 모델은 매 `2N+1`개 토큰마다 하나의 예측을 만듭니다.
예를 들어 `9`개 토큰 입력에 윈도우 크기 `2`면, `5`개의 예측을 반환합니다

## 2. 모델 (Model)

데이터가 준비되었으니 모델을 만들 차례입니다.
앞서 배운 커스텀 `nn.Module` 클래스 작성법을 활용해, 지금까지 배운 모든 것을 종합합니다.

모델 구조: **임베딩 → 은닉층(Linear + Tanh) → 출력층(Linear) → Sigmoid(확률)**


![model_architecture](assets/word_window_architecture.png)

```python
class WordWindowClassifier(nn.Module):

  def __init__(self, hyperparameters, vocab_size, pad_ix=0):
    super(WordWindowClassifier, self).__init__()

    """ 인스턴스 변수 """
    self.window_size = hyperparameters["window_size"]
    self.embed_dim = hyperparameters["embed_dim"]
    self.hidden_dim = hyperparameters["hidden_dim"]
    self.freeze_embeddings = hyperparameters["freeze_embeddings"]

    """ 임베딩 레이어
    임베딩 인덱스를 포함하는 텐서를 받아 해당하는 임베딩을 반환합니다.
    출력 차원은 (인덱스 수 * 임베딩 차원)입니다.

    freeze_embeddings가 True이면 임베딩 레이어 파라미터를 학습 불가능하게 설정합니다.
    이는 임베딩 파라미터 외의 파라미터만 변경하려는 경우에 유용합니다.

    """
    self.embeds = nn.Embedding(vocab_size, self.embed_dim, padding_idx=pad_ix)
    if self.freeze_embeddings:
      self.embeds.weight.requires_grad = False

    """ 은닉 레이어
    """
    full_window_size = 2 * window_size + 1
    self.hidden_layer = nn.Sequential(
      nn.Linear(full_window_size * self.embed_dim, self.hidden_dim),
      nn.Tanh()
    )

    """ 출력 레이어
    """
    self.output_layer = nn.Linear(self.hidden_dim, 1)

    """ 확률
    """
    self.probabilities = nn.Sigmoid()

  def forward(self, inputs):
    """
    B:= 배치 크기
    L:= 윈도우 패딩된 문장 길이
    D:= self.embed_dim
    S:= self.window_size
    H:= self.hidden_dim

    inputs: 토큰 인덱스의 (B, L) 텐서
    """
    B, L = inputs.size()

    """
    재구성.
    (B, L) LongTensor를 받아
    (B, L~, S) LongTensor를 출력합니다.
    """
    # 먼저, 입력의 각 단어에 대한 단어 윈도우를 가져옵니다.
    token_windows = inputs.unfold(1, 2 * self.window_size + 1, 1)
    _, adjusted_length, _ = token_windows.size()

    # 내부 텐서 크기 건전성 검사 (최소한 주석으로라도!)를 하는 것이 좋습니다!
    assert token_windows.size() == (B, adjusted_length, 2 * self.window_size + 1)

    """
    임베딩.
    크기가 (B, L~, S)인 torch.LongTensor를 받아
    (B, L~, S, D) FloatTensor를 출력합니다.
    """
    embedded_windows = self.embeds(token_windows)

    """
    재구성.
    (B, L~, S, D) FloatTensor를 받아
    (B, L~, S*D) FloatTensor로 크기를 조정합니다.
    -1 인수는 남은 축을 기반으로 마지막 차원이 무엇인지 "추론"합니다.
    """
    embedded_windows = embedded_windows.view(B, adjusted_length, -1)

    """
    첫 번째 레이어.
    (B, L~, S*D) FloatTensor를 받아
    (B, L~, H) FloatTensor로 크기를 조정합니다.
    """
    layer_1 = self.hidden_layer(embedded_windows)

    """
    두 번째 레이어
    (B, L~, H) FloatTensor를 받아
    (B, L~, 1) FloatTensor로 크기를 조정합니다.
    """
    output = self.output_layer(layer_1)

    """
    시그모이드.
    비정규화된 클래스 점수의 (B, L~, 1) FloatTensor를 받아
    (로그-)정규화된 클래스 점수의 (B, L~, 1) FloatTensor를 출력합니다.
    """
    output = self.probabilities(output)
    output = output.view(B, -1)

    return output
```

## 3. 학습 (Training)

이제 모든 것을 합칩니다. 데이터를 준비하고 모델을 초기화한 뒤,
옵티마이저와 손실 함수를 정의합니다.
이번에는 미리 정의된 손실 함수 대신 **직접 손실 함수를 정의**합니다.


```python
# 데이터 준비
data = list(zip(train_sentences, train_labels))
batch_size = 2
shuffle = True
window_size = 2
collate_fn = partial(custom_collate_fn, window_size=window_size, word_to_ix=word_to_ix)

# DataLoader 인스턴스화
loader = DataLoader(data, batch_size=batch_size, shuffle=shuffle, collate_fn=collate_fn)

# 모델 초기화
# 모든 모델 하이퍼파라미터를 딕셔너리에 넣는 것이 유용합니다.
model_hyperparameters = {
    "batch_size": 4,
    "window_size": 2,
    "embed_dim": 25,
    "hidden_dim": 25,
    "freeze_embeddings": False,
}

vocab_size = len(word_to_ix)
model = WordWindowClassifier(model_hyperparameters, vocab_size)

# 옵티마이저 정의
learning_rate = 0.01
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

# 이진 교차 엔트로피 손실을 계산하는 손실 함수 정의
def loss_function(batch_outputs, batch_labels, batch_lengths):
    # 전체 배치에 대한 손실 계산
    bceloss = nn.BCELoss()
    loss = bceloss(batch_outputs, batch_labels.float())

    # 손실 재조정. 각 훈련 예제에 있는 단어의 수를 저장하기 위해 길이를 사용했음을 기억하세요.
    loss = loss / batch_lengths.sum().float()

    return loss
```

이전 예제와 달리, 이번에는 매 에폭마다 전체 데이터를 한 번에 넣는 대신 **배치 단위로 반복**합니다.
따라서 각 학습 에폭 안에서 배치들을 순회합니다.


```python
# 매 에폭마다 호출될 함수
def train_epoch(loss_function, optimizer, model, loader):

  # 배치에 대한 총 손실을 추적합니다.
  total_loss = 0
  for batch_inputs, batch_labels, batch_lengths in loader:
    # 그래디언트 초기화
    optimizer.zero_grad()
    # forward pass 실행
    outputs = model.forward(batch_inputs)
    # 배치 손실 계산
    loss = loss_function(outputs, batch_labels, batch_lengths)
    # 그래디언트 계산
    loss.backward()
    # 파라미터 업데이트
    optimizer.step()
    total_loss += loss.item()

  return total_loss


# 메인 훈련 루프를 포함하는 함수
def train(loss_function, optimizer, model, loader, num_epochs=10000):

  # 각 에폭을 반복하고 train_epoch 함수를 호출합니다.
  for epoch in range(num_epochs):
    epoch_loss = train_epoch(loss_function, optimizer, model, loader)
    if epoch % 100 == 0: print(epoch_loss)
```

학습을 시작해봅시다!


```python
num_epochs = 1000
train(loss_function, optimizer, model, loader, num_epochs=num_epochs)
```

```python
# 테스트: 같은 'jordan'이 문맥에 따라 지역/사람으로 갈리는지 확인
# (학습 어휘집에 있는 단어만 사용 -> <unk> 없이 일반화 확인)
test_corpus = [
    "she traveled to jordan",           # jordan = 지역  -> 1 기대
    "michael jordan played basketball", # jordan = 사람  -> 0 기대
]
test_sentences = [s.lower().split() for s in test_corpus]
test_labels = [
    [0, 0, 0, 1],   # she traveled to [jordan]
    [0, 0, 0, 0],   # michael [jordan] played basketball
]

# 테스트 로더 생성
test_data = list(zip(test_sentences, test_labels))
collate_fn = partial(custom_collate_fn, window_size=2, word_to_ix=word_to_ix)
test_loader = torch.utils.data.DataLoader(test_data,
                                          batch_size=1,
                                          shuffle=False,
                                          collate_fn=collate_fn)
```

## 5. 보너스 — 임베딩이 학습한 *문맥* 들여다보기 🔍

`nn.Embedding`은 단어마다 **벡터 하나**를 줍니다 — 그래서 `jordan`의 임베딩 자체는 문맥과 무관하게 하나뿐입니다.
하지만 모델은 **윈도우(주변 단어)** 를 함께 본 뒤 은닉층에서 *문맥 표현(contextual representation)* 을 만듭니다.

Part 2에서 썼던 **PCA**로 이 표현을 2차원에 찍어보면, 같은 `jordan`이라도 문맥에 따라 어디에 놓이는지 눈으로 볼 수 있습니다.
(그림 안 글자는 폰트 문제로 영어로 표기합니다.)

### 🎬 시각 자료: 임베딩 학습 과정 — 문맥이 가르친다

같은 단어 jordan이 무작위로 흩어져 있다가, 학습이 진행되며 윈도우(문맥)가 지역 쓰임은 한쪽, 사람 쓰임은 다른 쪽으로 밀어냅니다 — 바로 아래 PCA 그림의 결과를 애니메이션으로 본 것입니다.

![EmbeddingTraining](assets/EmbeddingTraining.gif)

> ManimGL로 제작 · gif는 같은 저장소의 assets/ 에서 로드됩니다 (GitHub·로컬용)


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

pad_ix = word_to_ix["<pad>"]

# 한 문장의 각 단어에 대해, 그 단어의 윈도우를 은닉층에 통과시킨 "문맥 표현"을 구합니다.
def window_hidden(sent, model, S):
    idx = [pad_ix]*S + [word_to_ix.get(w, pad_ix) for w in sent] + [pad_ix]*S
    reps = []
    for i in range(len(sent)):
        emb = model.embeds(torch.LongTensor(idx[i:i+2*S+1])).reshape(-1)
        reps.append(model.hidden_layer(emb).detach().numpy())
    return reps

# 학습 문장들에서 'jordan'이 등장할 때마다 문맥 표현 + 지역/사람 레이블 수집
S = model.window_size
pts, labels, contexts = [], [], []
for sent, labs in zip(train_sentences, train_labels):
    for i, (w, lab, h) in enumerate(zip(sent, labs, window_hidden(sent, model, S))):
        if w == "jordan":
            pts.append(h)
            labels.append(lab)
            contexts.append(" ".join(sent[max(0, i-1):i+2]))

# PCA로 2차원 축소 후 산점도
XY = PCA(n_components=2).fit_transform(np.array(pts))
plt.figure(figsize=(7, 5))
for (x, y), lab, ctx in zip(XY, labels, contexts):
    plt.scatter(x, y, c="tab:green" if lab == 1 else "tab:red", s=140, edgecolors="k", zorder=3)
    plt.annotate(ctx, (x, y), fontsize=8, xytext=(5, 5), textcoords="offset points")
plt.title("'jordan' context representation  (green = LOCATION, red = person)")
plt.xlabel("PC1"); plt.ylabel("PC2"); plt.grid(alpha=0.3)
plt.tight_layout(); plt.show()

print("우리가 'jordan'의 뜻을 알려준 적은 없지만, 문맥(윈도우)만으로 지역/사람이 갈라집니다.")
```

### 윈도우 크기를 키우면?

윈도우가 클수록 **더 넓은 문맥**을 봅니다. 윈도우 크기를 `1 → 2 → 3`으로 바꿔 같은 그림을 그려보면,
두 의미(지역/사람)가 **점점 더 뚜렷하게 갈라지는** 것을 볼 수 있습니다 (각 그림의 `separation` 값이 커집니다).

```python
def train_ww(S, num_epochs=1000):
    global window_size
    window_size = S  # 모델 은닉층 크기가 이 전역 값을 참조하므로 함께 맞춰줍니다
    torch.manual_seed(0)
    data = list(zip(train_sentences, train_labels))
    loader = DataLoader(data, batch_size=2, shuffle=True,
                        collate_fn=partial(custom_collate_fn, window_size=S, word_to_ix=word_to_ix))
    hp = {"batch_size": 4, "window_size": S, "embed_dim": 25, "hidden_dim": 25, "freeze_embeddings": False}
    m = WordWindowClassifier(hp, len(word_to_ix))
    opt = torch.optim.SGD(m.parameters(), lr=0.01)
    for _ in range(num_epochs):
        for bx, by, bl in loader:
            opt.zero_grad()
            (loss_function(m.forward(bx), by, bl)).backward()
            opt.step()
    return m

def separation(XY, labs):
    XY, labs = np.array(XY), np.array(labs)
    gap = np.linalg.norm(XY[labs == 1].mean(0) - XY[labs == 0].mean(0))
    spread = (XY[labs == 1].std(0).mean() + XY[labs == 0].std(0).mean()) / 2
    return gap / (spread + 1e-6)

plt.figure(figsize=(13, 4))
for k, S in enumerate([1, 2, 3]):
    m = train_ww(S)
    pts, labs = [], []
    for sent, ls in zip(train_sentences, train_labels):
        for w, lab, h in zip(sent, ls, window_hidden(sent, m, S)):
            if w == "jordan":
                pts.append(h); labs.append(lab)
    XY = PCA(n_components=2).fit_transform(np.array(pts))
    ax = plt.subplot(1, 3, k + 1)
    for (x, y), lab in zip(XY, labs):
        ax.scatter(x, y, c="tab:green" if lab == 1 else "tab:red", s=110, edgecolors="k", zorder=3)
    ax.set_title(f"window_size = {S}   (separation {separation(XY, labs):.1f})")
    ax.grid(alpha=0.3)
plt.suptitle("Bigger window  ->  stronger separation of the two senses of 'jordan'")
plt.tight_layout(); plt.show()

window_size = model.window_size  # 원래 값(2)으로 복구
```
